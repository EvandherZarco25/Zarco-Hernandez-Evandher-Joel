## Resumen

El presente reporte detalla el desarrollo y los resultados obtenidos al crear un entorno virtual basado en el entorno físico localizado en el tercer piso del Centro de Ingeniería Avanzada (CIA). Este entorno virtual fue diseñado para entrenar a un robot en la tarea de manipulación de objetos, específicamente en agarrar y mover elementos de un lugar a otro. Se utilizaron herramientas de simulación avanzadas para replicar las condiciones físicas y optimizar el rendimiento del sistema robótico antes de implementar pruebas en el entorno real.


## Descripción

El proyecto consiste en el desarrollo de un banco de pruebas virtual para entrenar a un robot en la manipulación de objetos, replicando las condiciones del entorno físico localizado en el tercer piso del Centro de Ingeniería Avanzada (CIA). El entorno virtual fue diseñado para simular las características del espacio físico, incluyendo dimensiones, disposición de elementos, y las condiciones necesarias para realizar las pruebas de agarre y traslado de objetos.

Se utilizaron herramientas de simulación y programación, como Gazibo para modelar tanto el robot como el entorno. El robot fue configurado con sensores y actuadores simulados que permiten replicar su comportamiento real. El objetivo principal del entorno virtual es permitir pruebas y ajustes repetibles de manera segura, optimizando la eficiencia antes de implementar el sistema en el entorno físico.

Esta sección también incluye una descripción de los elementos clave del entorno físico, como las características del robot, los tipos de objetos a manipular, y las tareas específicas que el sistema debe realizar.

## Desarrollo

El proceso de creación del entorno virtual comenzó con la recopilación de datos clave del entorno físico localizado en el tercer piso del Centro de Ingeniería Avanzada (CIA). Estas mediciones incluyeron las dimensiones del espacio, la ubicación de los elementos relevantes y las características generales del ambiente.

Con estos datos, se utilizó la herramienta de creación de entornos de Gazebo para modelar el entorno virtual. Esta plataforma permitió replicar las dimensiones y disposiciones del espacio físico de manera precisa, asegurando que el entorno generado fuera lo más fiel posible a la realidad.

Una vez modelado el entorno, se generó el archivo .world, que contiene toda la información necesaria para su simulación en Gazebo. Este archivo representa el entorno virtual final y está listo para ser utilizado en pruebas de manipulación robótica, incluyendo la tarea de agarre y traslado de objetos.

Este desarrollo proporciona una base sólida para realizar simulaciones y entrenamientos de robots en un entorno seguro y controlado, previo a su implementación en el entorno físico real.

## Conclusiones

La creación del entorno virtual en Gazebo basado en las características del entorno físico del Centro de Ingeniería Avanzada (CIA) demostró ser una herramienta eficaz para el desarrollo y la validación de tareas robóticas. Este banco de pruebas virtual permitió replicar las condiciones reales con gran precisión, facilitando el entrenamiento del robot para manipular y trasladar objetos de manera segura y eficiente.

El uso de simulaciones no solo redujo los riesgos asociados a las pruebas físicas, sino que también optimizó el tiempo y los recursos necesarios para calibrar y ajustar el sistema. Este proyecto destaca la importancia de integrar tecnologías de simulación en la ingeniería robótica, mostrando su potencial para mejorar procesos, reducir costos y garantizar un diseño más robusto.

En general, el desarrollo realizado sienta una base sólida para futuras implementaciones y destaca el valor de la simulación como una herramienta clave en el diseño y prueba de soluciones robóticas.
